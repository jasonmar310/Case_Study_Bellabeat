```python
# Import required packages
import numpy as np
import matplotlib as plt
import pandas as pd
import datetime as dt
```

#### Import dataset

```python
data = pd.read_csv('dailyActivity_merged.csv')
```

#### Data exploration

```python
# View the data
data.head(10)
```

```python
data.info() # dtypes
```

```python
# Find ouy distinct Id of users
distinct_id = len(pd.unique(data['Id']))

print('Distinct of User ID: ' + str(distinct_id))
```

#### Data cleaning

```python
missing_values = data.isnull().sum()

print(f'Missing values in each column:\n{missing_values}')
```

```python
# Convert ActivityDate from string to datetime format
data['ActivityDate'] = pd.to_datetime(data['ActivityDate'])

# Check data type
data.dtypes
```

First, this dataset has no missing or null values. Data frame includes 940 rows and 15 columns, and there are 3 unique ID from fitness tracker users.

Also, we convert ActivityDate to datetime64 data type (yyyy-mm-dd)

```python
# Seperate the date into day of the week as a new column
data['DayOfTheWeek'] = data['ActivityDate'].dt.day_name()

data['DayOfTheWeek'].head(10)
```

```python
# Rename columns
data.rename(columns = {"Id":"id", "ActivityDate":"date", "DayOfTheWeek":"day_of_the_week", 
                       "TotalSteps":"total_steps", "TotalDistance":"total_dist", "TrackerDistance":"track_dist", 
                       "LoggedActivitiesDistance":"logged_dist", "VeryActiveDistance":"very_active_dist", 
                       "ModeratelyActiveDistance":"moderate_active_dist", "LightActiveDistance":"light_active_dist", 
                       "SedentaryActiveDistance":"sedentary_active_dist", "VeryActiveMinutes":"very_active_mins", 
                       "FairlyActiveMinutes":"fairly_active_mins", "LightlyActiveMinutes":"lightly_active_mins", 
                       "SedentaryMinutes":"sedentary_mins", "TotalExerciseMinutes":"total_mins","TotalExerciseHours":"total_hours",
                       "Calories":"calories"}, inplace = True)

data.head(5)
```

```python
# Create new column 'minutes' containing sum of total minutes
data['minutes'] = data['very_active_mins'] + data['fairly_active_mins'] + data['lightly_active_mins'] + data['sedentary_mins']

data['minutes'].head(10)
```

```python
# Create a 'hour' column by transforming the minutes column
data['hours'] = round(data['minutes']/60)

data['hours'].head(10)
```

### Analyze


4.1 Perform calculations


```python
# General statistics
data.describe()
```

```python
# Calculate average steps, distance, and active minutes for each day of the week
average_steps = data.groupby('day_of_the_week')['total_steps'].mean()
average_distance = data.groupby('day_of_the_week')['total_dist'].mean()
average_active_minutes = data.groupby('day_of_the_week')['minutes'].mean()

# Display the results
print("Average Steps by Day of Week:\n", average_steps)
print("\nAverage Distance by Day of Week:\n", average_distance)
print("\nAverage Active Minutes by Day of Week:\n", average_active_minutes)
```

```python
# Calculate the average values
average_values = data[['total_steps', 'total_dist', 'very_active_dist', 'moderate_active_dist', 'light_active_dist', 'minutes']].mean()

# Calculate the number of users who have above-average values
frequent_users = (data[['total_steps', 'total_dist', 'very_active_dist', 'moderate_active_dist', 'light_active_dist', 'minutes']] > average_values).sum()

# Display the number of frequent users for each metric
print("Number of Frequent Users for Each Metric:\n", frequent_users)
```

#### In-depth analysis


1. Time-Series Analysis


2. Correltation Analysis


3. User Segmentation


```python
# Time-series analysis
data_copy = data.copy()

# Set the date column as the index in the copy
data_copy.set_index('date', inplace=True)

# Plot the average number of steps per month
data_copy['total_steps'].resample('M').mean().plot()
```

```python
# Correlation analysis
correlation = data[['total_steps', 'total_dist', 'very_active_dist', 'moderate_active_dist', 'light_active_dist', 'calories']].corr()

print('Correlation: ', correlation)

# Identify variables with high correlation (above 0.7)
high_correlation = correlation[correlation > 0.7]

high_correlation[high_correlation == 1] = np.nan # exclude self-correlation

print("\n\nHigh correlation: ", high_correlation)
```

```python
# Based on the high correlation, we will build a regression model to verify the variables

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

# Assume we are using 'total_steps' to predict 'total_dist'
X = data['total_steps'].values.reshape(-1,1)
y = data['total_dist'].values.reshape(-1,1)

# Split the data into train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Linear regression model
regressor = LinearRegression()

regressor.fit(X_train, y_train)

# Make predictions on the test data
y_pred = regressor.predict(X_test)

# Compare the predicted values to the actual values
print('Mean Absolute Error:', round(metrics.mean_absolute_error(y_test, y_pred), 4))
print('Mean Squared Error:', round(metrics.mean_squared_error(y_test, y_pred), 4))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
# User Segmentation

from sklearn.cluster import KMeans
import seaborn as sns

# Select relevant columns
X = data[['total_steps', 'total_dist', 'very_active_dist', 'moderate_active_dist', 'light_active_dist', 'calories']]

# Create a KMeans model with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)

data['cluster'] = kmeans.labels_

# Descriptive statistics for each cluster
cluster_stats = data.groupby('cluster').mean()

print(cluster_stats)
```

```python
# Calculating total of individual minutes column
very_active_mins = data["very_active_mins"].sum()
fairly_active_mins = data["fairly_active_mins"].sum()
lightly_active_mins = data["lightly_active_mins"].sum()
sedentary_mins = data["sedentary_mins"].sum()

# Pie chart
slices = [very_active_mins, fairly_active_mins, lightly_active_mins, sedentary_mins]
labels = ["Very active minutes", "Fairly active minutes", "Lightly active minutes", "Sedentary minutes"]
colours = ["darkorange", "gold", "darkgreen", "lightgreen"]
explode = [0, 0, 0, 0.1]
plt.style.use("default")
plt.pie(slices, labels = labels, 
        colors = colours, wedgeprops = {"edgecolor": "black"}, 
        explode = explode, autopct = "%1.1f%%")
plt.title("Percentage of Activity in Minutes")
plt.tight_layout()
plt.show()
```

#### Percentage of activity in minutes

1. The majority (81.3%) of logged minutes are categorized as "Sedentary", indicating that users primarily employ the FitBit app to monitor non-exercise activities such as commuting or routine movements.


2. The small proportion of "Fairly Active" (1.1%) and "Very Active" (1.7%) minutes suggests that users are not extensively utilizing the FitBit app for tracking fitness activities like running or high-intensity workouts.


3. This underutilization of FitBit for fitness tracking contrasts with the app's central objective of encouraging and monitoring physical fitness, indicating a potential area for user engagement improvement.

```python
app = dash.Dash(__name__)

# Create a line graph for average steps
avg_steps_fig = go.Figure()
avg_steps_fig.add_trace(
    go.Scatter(x=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], 
               y=average_steps.tolist(),
               mode='lines+markers',
               name='Average Steps'))
avg_steps_fig.update_layout(
    title="Average Steps by Day of the Week",
    xaxis_title="Day of the Week",
    yaxis_title="Average Steps",
)

# Create a bar chart for average distance
avg_dist_fig = go.Figure()
avg_dist_fig.add_trace(
    go.Bar(x=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], 
            y=average_distance.tolist(),
            name='Average Distance'))
avg_dist_fig.update_layout(
    title="Average Distance by Day of the Week",
    xaxis_title="Day of the Week",
    yaxis_title="Average Distance",
)

# Create a pie chart for activity distribution
activity_dist_fig = go.Figure(
    go.Pie(labels=['Very Active', 'Fairly Active', 'Lightly Active', 'Sedentary'], 
           values=[very_active_mins, fairly_active_mins, lightly_active_mins, sedentary_mins],
           name='Activity Distribution'))
activity_dist_fig.update_layout(
    title="Activity Distribution in Minutes",
)

app.layout = html.Div(children=[
    html.H1(children='Fitbit Data Dashboard'),

    html.Div(children=[
        dcc.Graph(
            id='average-steps',
            figure=avg_steps_fig
        ),
        dcc.Graph(
            id='average-distance',
            figure=avg_dist_fig
        ),
        dcc.Graph(
            id='activity-distribution',
            figure=activity_dist_fig
        ),
    ])
])

if __name__ == '__main__':
    app.run_server(debug=True)
```
